```python
import numpy as np
import pandas as pd 

from sklearn.model_selection import train_test_split
from sklearn import preprocessing

# Importing plotting libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datafile = 'Resources/IowaData.csv'
df = pd.read_csv(datafile)

df.head()
```

```python
expensive = df.loc[df["SalePrice"] > 350000]
# new_index = expensive.reset_index(drop=True)
expensive.head()
```

```python
cols = [c for c in expensive.columns]
```

```python
f, ax = plt.subplots(figsize=(25, 15))
corr = expensive[cols].corr()


hm = sns.heatmap(round(corr,2), annot=False, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05) 
f.subplots_adjust(top=0.93)
t= f.suptitle('Housing Attributes Correlation Heatmap', fontsize=18)
```
